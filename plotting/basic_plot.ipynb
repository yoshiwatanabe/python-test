# Line plot
Use line plot to show how one variable changes based the other on some intervals.

```python
import matplotlib.pyplot as plt
import numpy as np

# Sample data
x = np.linspace(0, 10, 100)
y = np.sin(x)

# Create line plot
plt.figure(figsize=(8, 5))
plt.plot(x, y, label='Sine Wave')
plt.title('Line Plot')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.legend()
plt.show()
```

# Scatter plot
Use scatter plot to show relationship between two variables. It can show if two variables have some correlation or just random. Points would form some grouping that would intuively indicate relationship. It can show clusters. It can also show a function.

```python
import matplotlib.pyplot as plt
import numpy as np

# Sample data
x = np.random.rand(50)
y = np.random.rand(50)

# Create scatter plot
plt.figure(figsize=(8, 5))
plt.scatter(x, y, color='green')
plt.title('Scatter Plot')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.show()
```

# Bar chart
Use bar chart to compare values among categories. The size of bars intuitively represents the relative relationships between categories.

```python
import matplotlib.pyplot as plt

# Sample data
categories = ['A', 'B', 'C', 'D', 'E']
values = [5, 7, 3, 8, 4]

# Create bar chart
plt.figure(figsize=(8, 5))
plt.bar(categories, values, color='skyblue')
plt.title('Bar Chart')
plt.xlabel('Categories')
plt.ylabel('Values')
plt.show()
```

# Pie chart
Useful to show proportions of different categories.

```python
import matplotlib.pyplot as plt

# Sample data
labels = ['Apple', 'Banana', 'Cherry', 'Date']
sizes = [15, 30, 45, 10]

# Create pie chart
plt.figure(figsize=(6, 6))
plt.pie(sizes, labels=labels, autopct='%1.1f%%', startangle=90)
plt.title('Pie Chart')
plt.show()
```

# Box plot
Box plots display the distribution of data based on a five-number summary: minimum, first quartile, median, third quartile, and maximum.

```python
import matplotlib.pyplot as plt
import numpy as np

# Sample data
data = [np.random.normal(0, std, 100) for std in range(1, 4)]

# Create box plot
plt.figure(figsize=(8, 5))
plt.boxplot(data, labels=['Std 1', 'Std 2', 'Std 3'])
plt.title('Box Plot')
plt.xlabel('Dataset')
plt.ylabel('Value')
plt.show()
```

# Violin plot
Violin plots are similar to box plots but with a rotated kernel density plot on each side.

```python
import matplotlib.pyplot as plt
import numpy as np

# Sample data
data = [np.random.normal(0, std, 100) for std in range(1, 4)]

# Create violin plot
plt.figure(figsize=(8, 5))
plt.violinplot(data, showmedians=True)
plt.title('Violin Plot')
plt.xlabel('Dataset')
plt.ylabel('Value')
plt.xticks([1, 2, 3], ['Std 1', 'Std 2', 'Std 3'])
plt.show()
```

# Head mapp

# Contour plot
Contour plots are used to represent three-dimensional data in two dimensions using contour lines.

```python
import matplotlib.pyplot as plt
import numpy as np

# Sample data
x = np.linspace(-3, 3, 100)
y = np.linspace(-3, 3, 100)
X, Y = np.meshgrid(x, y)
Z = np.sin(X**2 + Y**2)

# Create contour plot
plt.figure(figsize=(8, 6))
plt.contour(X, Y, Z, levels=20, cmap='plasma')
plt.title('Contour Plot')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.colorbar()
plt.show()
```

# 3D surface plot
3D plots allow you to visualize data in three dimensions.

```python
import matplotlib.pyplot as plt
import numpy as np
from mpl_toolkits.mplot3d import Axes3D

# Sample data
x = np.linspace(-5, 5, 50)
y = np.linspace(-5, 5, 50)
X, Y = np.meshgrid(x, y)
Z = np.sin(np.sqrt(X**2 + Y**2))

# Create 3D plot
fig = plt.figure(figsize=(8, 6))
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(X, Y, Z, cmap='viridis')
ax.set_title('3D Surface Plot')
ax.set_xlabel('X-axis')
ax.set_ylabel('Y-axis')
ax.set_zlabel('Z-axis')
plt.show()
```

# Error bar plot

```python
import matplotlib.pyplot as plt
import numpy as np

# Sample data
x = np.arange(0, 10, 1)
y = np.random.normal(5, 0.5, size=len(x))
error = np.random.uniform(0.2, 0.4, size=len(x))

# Create error bar plot
plt.figure(figsize=(8, 5))
plt.errorbar(x, y, yerr=error, fmt='o', ecolor='red', capsize=5)
plt.title('Error Bar Plot')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.show()
```

# Stack plot
Stack plots show different datasets stacked on top of each other.

```python
import matplotlib.pyplot as plt
import numpy as np

# Sample data
days = [1, 2, 3, 4, 5]
sleeping = [7, 8, 6, 11, 7]
eating = [2, 3, 4, 3, 2]
working = [7, 8, 7, 2, 2]
playing = [8, 5, 7, 8, 13]

# Create stack plot
plt.figure(figsize=(8, 5))
plt.stackplot(days, sleeping, eating, working, playing, labels=['Sleeping', 'Eating', 'Working', 'Playing'])
plt.title('Stack Plot')
plt.xlabel('Days')
plt.ylabel('Hours')
plt.legend(loc='upper left')
plt.show()
```

# Polar plot
Polar plots represent data in polar coordinates.

```python
import matplotlib.pyplot as plt
import numpy as np

# Sample data
theta = np.linspace(0, 2*np.pi, 100)
r = np.abs(np.sin(5*theta))

# Create polar plot
plt.figure(figsize=(6, 6))
ax = plt.subplot(111, projection='polar')
ax.plot(theta, r)
ax.set_title('Polar Plot')
plt.show()
```

# Quiver plot

```python
import matplotlib.pyplot as plt
import numpy as np

# Sample data
x = y = np.arange(-2, 2, 0.2)
X, Y = np.meshgrid(x, y)
U = -Y
V = X

# Create quiver plot
plt.figure(figsize=(7, 7))
plt.quiver(X, Y, U, V)
plt.title('Quiver Plot')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.show()
```

# Stream plot

```python
import matplotlib.pyplot as plt
import numpy as np

# Sample data
Y, X = np.mgrid[-3:3:100j, -3:3:100j]
U = -1 - X**2 + Y
V = 1 + X - Y**2

# Create stream plot
plt.figure(figsize=(8, 6))
plt.streamplot(X, Y, U, V, density=1.5, color='blue')
plt.title('Stream Plot')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.show()
```
