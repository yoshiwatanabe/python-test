# 最小二乗法

## 最小二乗法の可視化

[インターアクティブに最小二乗法を試すページを開く](https://yoshiwatanabe.github.io/python-test/least_squires_method.html)

インターアクティブページを開くと、以下のような状態になっているはずです。
ここでは、いくつかのデータポイントが散布図に配置してあり、さらに$y=ax + b$という直線に対して書くデータポイントからの誤差を（誤差の２乗）としての「正方形（赤枠）」で表しています。

さらに、それらの正方形をすべて足し合わせた「合計の正方形（薄緑色）」がオーバーレイ表示されているのが分かります。最小二乗法はこの「合計の正方形が最も小さくなるように」$y=ax + b" の$a$ と $b$、すなわち傾きと切片を調整して「フィットさせる」ことです。


傾きと切片をスライダーで調整して、以下のように「合計の正方形」が出来るだけ小さくなる組み合わせを探します。
