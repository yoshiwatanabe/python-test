# 統計的仮説検定


## 基本的な流れ
1. 帰無仮説（$H_0$）と対立仮説（$H_1$）を設定する
   - 帰無仮説（$H_0$）：「差はない」「効果はない」「偶然である」という仮説  
   - 対立仮説（$H_1$）：「差がある」「効果がある」「偶然ではない」という仮説

2. データを収集し、統計量を計算する
   - 平均や標準偏差を求め、適切な検定手法（t検定、χ²検定など）を用いて計算する。

3. 有意水準（$\alpha$）を決める
   - 一般的には $ \alpha = 0.05 $（5%） を使用。
   - これは「誤って帰無仮説を棄却する確率（第1種の過誤）」を示す。

4. p値（p-value）を算出する
   - p値は「帰無仮説が正しいと仮定したときに、観測データ以上の結果が得られる確率」。
   - p値が 有意水準（$\alpha$）より小さければ、帰無仮説を棄却する。

5. 結論を出す
   - p値 < 0.05 → 帰無仮説を棄却し、「差がある」と判断する。
   - p値 ≥ 0.05 → 帰無仮説を棄却できず、「差があるとは言えない」と判断する。


### 具体例：新しい薬の効果を検証
仮説設定
- $H_0$（帰無仮説）：「新しい薬と従来の薬の効果に差はない」
- $H_1$（対立仮説）：「新しい薬の方が効果が高い」

データ収集
- 100人ずつの2グループ（新薬 vs. 旧薬）で比較。
- それぞれのグループの平均回復時間を計算。

t検定を実施 → p値 = 0.03
- p値 < 0.05 なので、帰無仮説を棄却 → 「新薬は旧薬より効果がある」と結論。
