# 中心極限定理

インターアクティブ・シミュレーターを使って、「独立した確率変数の和または平均」が正規分布に近づくようすを実験しましょう。

[中心極限定理シミュレーションのページを開く](https://yoshiwatanabe.github.io/python-test/central_limit_theorem.html)

（プレビュー）


## 解説

1. **中心極限定理の条件**:
   - 中心極限定理は「独立した確率変数の**和または平均**」が正規分布に近づくことを示す定理です
   - 重要なのは「複数のランダム変数」の平均を取ることです

2. **サンプルサイズ1の意味**:
   - サンプルサイズ1とは、各試行で1つだけの乱数を生成し、その値をそのまま使うことを意味します
   - この場合、平均を取る操作が発生しないため、元の分布の形状がそのまま維持されます

3. **0〜1の一様分布の特性**:
   - サンプルサイズ1の場合、結果は元の確率分布（この場合は一様分布）に従います
   - 試行回数を増やしても、各試行は独立しており、データ点が増えるだけで分布の形状は変わりません

4. **試行回数とサンプルサイズの違い**:
   - 試行回数は「何回実験を繰り返すか」（ヒストグラムの精度に影響）
   - サンプルサイズは「1回の実験で何個のデータを平均するか」（分布の形状に影響）

簡単に例えると：
- サンプルサイズ1：1回のサイコロを振って結果を記録。これを何回繰り返しても、サイコロの目の分布は変わらない
- サンプルサイズ5：5回サイコロを振って平均値を記録。これを繰り返すと、記録された平均値は正規分布に近づく

つまり、中心極限定理が適用されるためには、サンプルサイズが1より大きい必要があり、サンプルサイズが大きくなるほど正規分布への近似が良くなります。試行回数を増やすことは分布の推定精度を上げるだけで、分布の形状自体は変化させません。
