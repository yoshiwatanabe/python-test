# 足し算と引き算を一次関数として統一的に理解する

四則演算のうちの足し算と引き算、つまり「和」と「差」は一次関数として捉えて統一的に理解できます。

一次関数（linier function）は、与えられた値とその結果をグラフで表すと「直線」になります。一次関数の一般的な形は$y=kx+b$となり、$b$ の値の違いが「足し算」と「引き算」の違いになります。（引き算も実は足し算、というところは今は無視しましょう）

## 足し算

足し算は、何かの数値に「３を足す」や「それに３を加える」という意味ですが、その性質をグラフで表すとどうなるのでしょうか？

$k$については今は脇に置いておくため$1$だとして、以下のグラフの特徴を見てみます。

```python
# 必要なライブラリをインポート
import numpy as np
import matplotlib.pyplot as plt

# 係数を固定（例として k = 1 に設定）
k = 1

# 定数項（足し算・引き算）としての b のリスト
constants = [0, 3, 5]

# x の値を生成
x_vals = np.linspace(-10, 10, 400)

# グラフを描画
plt.figure(figsize=(8, 6))

# 各定数項 b についてループ
for b in constants:
    # 一次関数を定義
    y_vals = k * x_vals + b
    # ラベルを設定
    label = f'y = {k}x + {b}'
    # グラフを描画
    plt.plot(x_vals, y_vals, label=label)

# グラフの設定
plt.xlabel('x')
plt.ylabel('y')
plt.grid(True)
plt.legend()
plt.show()
```

- 定数項が$0$の時：$x$がどんな値でも$y$は常に$x$の項の結果（何も足さないし、何も引かない・・・「天気の子」みたい）
- 定数項が$3$の時：$x$がどんな値でも$y$は$x$の項の結果に$3$を足した値
- 定数項が$5$の時：$x$がどんな値でも$y$は$x$の項の結果に$5$を足した値

つまり「足し算」は「$x$の項の結果」を一定の値分増加させる処理だと理解できます。

# 引き算

引き算についても以下のグラフの特徴を見てみます。

```python
# 必要なライブラリをインポート
import numpy as np
import matplotlib.pyplot as plt

# 係数を固定（例として k = 1 に設定）
k = 1

# 定数項（足し算・引き算）としての b のリスト
constants = [-5, -3, 0]

# x の値を生成
x_vals = np.linspace(-10, 10, 400)

# グラフを描画
plt.figure(figsize=(8, 6))

# 各定数項 b についてループ
for b in constants:
    # 一次関数を定義
    y_vals = k * x_vals + b
    # ラベルを設定
    label = f'y = {k}x + {b}'
    # グラフを描画
    plt.plot(x_vals, y_vals, label=label)

# グラフの設定
plt.xlabel('x')
plt.ylabel('y')
plt.grid(True)
plt.legend()
plt.show()
```

- 定数項が$0$の時：$x$がどんな値でも$y$は常に$x$の項の結果（何も足さないし、何も引かない・・・「天気の子」みたい）
- 定数項が$-3$の時：$x$がどんな値でも$y$は$x$の項の結果から$3$を引いた値
- 定数項が$-5$の時：$x$がどんな値でも$y$は$x$の項の結果から$5$を引いた値

つまり「引き算」は「$x$の項の結果」を一定の値分減少させる処理だと理解できます。

# 抽象化
足し算を考えるとき、まずは $10 + 3$ のような具体的な数値で考えることから始めます。次に、$+3$ に着目すると「ある数に $3$ を足す」という演算操作が見えてきます。未知の値を代入したとき、結果は「その値に $3$ を加えたもの」になります。

抽象化して表すと、一次関数などの関数の定数項として表現されるわけです。ここで、定数項は正の値でも負の値でもあり得ます。例えば、一次関数 $y = x + 3$ は「$x$ に $3$ を足す」操作を表し、$y = x - 2$ は「$x$ から $2$ を引く」操作を表します。
