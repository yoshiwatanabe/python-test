### 三角関数 $sin(x)$, $cos(x)$, $tan(x)$ の特徴

３つの三角関数の特徴をまとめると：

| **関数**     | **特徴**                                                                 | **単位円での解釈**                                                                 |
|--------------|--------------------------------------------------------------------------|------------------------------------------------------------------------------------|
| $sin(x)$  | 周期: $2\pi$, 振幅: 1, 波形: 上下対称の波形 ($[-1, 1]$)             | 単位円の **y座標** を表す。角度 $x$ に対し、点の高さが $sin(x)$ の値となる。  |
| $cos(x)$  | 周期: $2\pi$, 振幅: 1, 波形: 左右対称の波形 $[-1, 1]$             | 単位円の **x座標** を表す。角度 $x$ に対し、点の横の位置が $cos(x)$ の値。   |
| $tan(x)$  | 周期: $\pi$, 無限に伸びる波形（漸近線あり）。                          | 単位円で $sin(x)/cos(x)$ を表す。角度 $x$ に対し、x座標の比で決定される。    |

波の特徴：

| **関数**     | **特徴**                                                                 |
|--------------|--------------------------------------------------------------------------|
| $sin(x)$  | お椀を伏せたような曲線が、お椀の左端を$(0,0)$に合わせた形で始まる。そして同じ形のお椀をひっくり返したものが続く。これは時計の針が３を指している状態からスタートして、反時計回りにぐるっと一周した状態に対応する。針が３時の時の$y$(高さ)は$0$。12時の時は$y$（高さ）は最大の$1$（単位円として）。この要領で針が指し示す時計の円周上に赤い点があると想像し、それに対応する角度を$0$から$2\pi$まで$sin(x)$に渡した際の$x$と$y$の点をプロットして出来た曲線のイメージ          |
| $cos(x)$  | お椀を伏せて真っ二つになったような曲線が、切った中心をスタートにして始まる。そして同じ形のお椀をひっくり返したものを真っ二つに切ったものに続き、さらにその残りが続く。最後に最初のお椀の残りに続く。これは時計の針が３を指している状態からスタートして、反時計回りにぐるっと一周した状態に対応する。針が３時の時の$x$(横の幅)は最大の$1$。12時の時は$x$（横の幅）は$0$になる。この要領で針が指し示す時計の円周上に赤い点があると想像し、それに対応する角度を$0$から$2\pi$まで$sin(x)$に渡した際の$x$と$y$の点をプロットして出来た曲線のイメージ          |
| $tan(x)$  | 時計のイメージを使うと、３時のスタート地点では角度が$0$なので$Sin(0)$と$Cos(0)$はそれぞれ$0$と$1$になる(高さ$0$、横$1$)。$tan(x)$は$Sin(0)$を$Cos(0)$で割ったものなので$0$になる。途中で高さと横が同じ値になるところがある。ここで$y=1$になる。そこを過ぎて、時計の針が１２時近くまで分母はどんどん小さくなり１２時直前では分母がめちゃくちゃ小さいので高さはめちゃくちゃおおきくなる（無限大）。１２時から９時の間は逆に$y＝oo$でスタートして、分母がどんどん増えて$0$になる。ここからまたもういちど半周期がはじまる                 |



# 三角関数と単位円の関係

三角関数という関数は「角度」を入力とし、その角度の時の「高さ」や「横の長さ」や「縦横の比率」を出力してくれます。
つまり

- 角度＝＞縦の長さ
- 角度＝＞横の長さ
- 角度＝＞縦横の長さの比

になる関数ということです。

この三つの関数は「単位円」という特別な円で考えると分かりやすい。特別な円と言っても簡単で、円の半径が１の円だということ。比率については「割る数」が１だとすごく単純化できるので、わざわざ半径を１にした円を使っている。



```python

import numpy as np
import matplotlib.pyplot as plt

# Define the circle
theta = np.linspace(0, 2 * np.pi, 500)  # Full circle angles
circle_x = np.cos(theta)
circle_y = np.sin(theta)

# Define a specific angle to demonstrate the relationship
angle = np.pi / 4  # Example: 45 degrees (π/4 radians)
x_point = np.cos(angle)
y_point = np.sin(angle)

# Create the plot
fig, ax = plt.subplots(figsize=(8, 8))

# Draw the unit circle
ax.plot(circle_x, circle_y, label="Unit Circle", color="blue")

# Highlight the specific angle
ax.plot([0, x_point], [0, y_point], linestyle="--", color="orange", label="Radius (angle = π/4)")
ax.scatter([x_point], [y_point], color="red", label=f"Point (x={x_point:.2f}, y={y_point:.2f})")

# Draw projections onto the x-axis and y-axis
ax.plot([x_point, x_point], [0, y_point], linestyle=":", color="green", label="Projection on y-axis")
ax.plot([0, x_point], [y_point, y_point], linestyle=":", color="purple", label="Projection on x-axis")

# Annotate the angle and projections
ax.annotate(f"Angle = π/4\nx = cos(π/4)\ny = sin(π/4)", xy=(x_point / 2, y_point / 2), color="red")

# Set axis limits and aspect ratio
ax.set_xlim(-1.5, 1.5)
ax.set_ylim(-1.5, 1.5)
ax.set_aspect('equal', adjustable='box')

# Add grid, labels, and legend
ax.axhline(0, color='black', linewidth=0.5, linestyle='--')
ax.axvline(0, color='black', linewidth=0.5, linestyle='--')
ax.grid(True)
ax.legend()
ax.set_title("Unit Circle and Relationship Between sin(x), cos(x), and Angle")
ax.set_xlabel("x")
ax.set_ylabel("y")

# Show the plot
plt.show()
```

単位円を使って考えると

- Sine関数は、縦の長さを１で割るので縦の逃さそのもの
- Cosine関数は、横の長さを１で割るので横の長さそのもの
- Tangent関数は、単位円の半径１は関係なくて、縦の長さを横の長さで割ったもの


平たく言うと

- Sine関数は、ある角度の時の高さを表す
- Cosine関数は、ある角度の時の横の長さを表す
- Tangent関数は、ある角度の時のSine関数とCosine関数の比率を表す


# ラジアンで表す角度のおさらい

*1ラジアン*は、弧の長さが円の半径と等しいときの中心角です。実際には約57.3度ぐらいの角度になります。
以下の図の「赤い点」の座標は$Sin(1)$と$Cos(1)$で求めることが出来ます。

この時の角度の「単位」はラジアン。

```python
# Re-importing libraries after reset
import numpy as np
import matplotlib.pyplot as plt

# Define the circle and 1 radian arc
theta = np.linspace(0, 2 * np.pi, 500)  # Full circle angles
circle_x = np.cos(theta)
circle_y = np.sin(theta)

# Define 1 radian arc
radian_angle = 1  # 1 radian
arc_theta = np.linspace(0, radian_angle, 100)
arc_x = np.cos(arc_theta)
arc_y = np.sin(arc_theta)

# Create the plot
fig, ax = plt.subplots(figsize=(8, 8))

# Draw the unit circle
ax.plot(circle_x, circle_y, label="Unit Circle", color="blue")

# Highlight the 1 radian arc
ax.plot(arc_x, arc_y, color="orange", label="1 Radian Arc")
ax.scatter([1, np.cos(radian_angle)], [0, np.sin(radian_angle)], color="red", zorder=5)
ax.annotate("Start (1, 0)", xy=(1, 0), xytext=(1.1, -0.1), color="red")
ax.annotate(f"End ({np.cos(radian_angle):.2f}, {np.sin(radian_angle):.2f})", 
            xy=(np.cos(radian_angle), np.sin(radian_angle)), 
            xytext=(np.cos(radian_angle) + 0.2, np.sin(radian_angle) + 0.1), 
            color="red")

# Add lines for the radius
ax.plot([0, 1], [0, 0], linestyle="--", color="green", label="Radius to Start")
ax.plot([0, np.cos(radian_angle)], [0, np.sin(radian_angle)], linestyle="--", color="purple", label="Radius to End")

# Set axis limits and aspect ratio
ax.set_xlim(-1.5, 1.5)
ax.set_ylim(-1.5, 1.5)
ax.set_aspect('equal', adjustable='box')

# Add grid, labels, and legend
ax.axhline(0, color='black', linewidth=0.5, linestyle='--')
ax.axvline(0, color='black', linewidth=0.5, linestyle='--')
ax.grid(True)
ax.legend()
ax.set_title("1 Radian on the Unit Circle")
ax.set_xlabel("x")
ax.set_ylabel("y")

# Show the plot
plt.show()
```


ラジアン（radian）は、円の中心から円周上の点に向かって引いた線が作る角度を、**円周の長さに基づいて**表す単位です。具体的には、ラジアンは次のように定義されます：

1. 円の半径を $r$ とします。
2. 円周上の2点間の**弧の長さ**を $s$ とします。
3. この弧が作る中心角を $\theta$ とします。

このとき、ラジアンは次の式で表されます：

$$
\theta = \frac{s}{r}
$$

円周の全長は $2\pi r$ なので、円周全体に対応する角度は $2\pi$ ラジアンとなります。

したがって、ラジアンは**弧の長さを半径で割った比率**であり、円周率 $\pi$ を用いて角度を表すことができます。
