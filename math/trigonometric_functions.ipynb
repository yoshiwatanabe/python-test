# Sin,Cos,Tan関数のイメージ

１次関数、２次関数のような関数は周期をもたないけれど、三角関数は周期関数だということがまうひとつ。

次に、それぞれのの関数の全体的な形を見て特徴を理解する。

```python
import numpy as np
import matplotlib.pyplot as plt

# Define x values (from -2π to 2π for better visualization)
x = np.linspace(-2 * np.pi, 2 * np.pi, 1000)

# Define the functions
y_sin = np.sin(x)  # Sine function
y_cos = np.cos(x)  # Cosine function
y_tan = np.tan(x)  # Tangent function

# Limit y values for tangent to avoid extreme values (visual clarity)
y_tan[np.abs(y_tan) > 10] = np.nan

# Create the plot with aspect ratio 1:1 and x-axis labels every 5 intervals
plt.figure(figsize=(12, 6))

# Plot sine
plt.plot(x, y_sin, label="sin(x)", color="blue")

# Plot cosine
plt.plot(x, y_cos, label="cos(x)", color="orange")

# Plot tangent
plt.plot(x, y_tan, label="tan(x)", color="green", linestyle="--")

# Add horizontal and vertical lines
plt.axhline(0, color='black', linewidth=0.5)
plt.axvline(0, color='black', linewidth=0.5)

# Set x-axis ticks and labels at every 5 interval
plt.xticks(np.arange(-10, 11, 5))  # Adjusting for a more realistic scale
plt.yticks(np.arange(-3, 4, 1))  # Y-axis labels remain as integers

# Add grid lines
plt.grid(True, which='both', linestyle='--', linewidth=0.5)

# Add labels and legend
plt.title("Graphs of sin(x), cos(x), and tan(x) with X-axis Labels Every 5 Intervals")
plt.xlabel("x")
plt.ylabel("y")
plt.legend()

# Set aspect ratio to 1:1
plt.gca().set_aspect('equal', adjustable='box')

# Show the plot
plt.show()
```

三角関数 $sin(x)$, $cos(x)$, $tan(x)$ の特徴をまとめると

| **関数**     | **特徴**                                                                 | **単位円での解釈**                                                                 |
|--------------|--------------------------------------------------------------------------|------------------------------------------------------------------------------------|
| $sin(x)$  | 周期: $2\pi$, 振幅: 1, 波形: 上下対称の波形 ($[-1, 1]$)             | 単位円の **y座標** を表す。角度 $x$ に対し、点の高さが $sin(x)$ の値となる。  |
| $cos(x)$  | 周期: $2\pi$, 振幅: 1, 波形: 左右対称の波形 $[-1, 1]$             | 単位円の **x座標** を表す。角度 $x$ に対し、点の横の位置が $cos(x)$ の値。   |
| $tan(x)$  | 周期: $\pi$, 無限に伸びる波形（漸近線あり）。                          | 単位円で $sin(x)/cos(x)$ を表す。角度 $x$ に対し、x座標の比で決定される。    |

よりくだけたイメージには時計のような円と角度を表すものの状態を考えながらそれぞれの関数の特徴を捉えることができる

| **関数**     | **特徴**                                                                 |
|--------------|--------------------------------------------------------------------------|
| $sin(x)$  | お椀を伏せたような曲線が、お椀の左端を$(0,0)$に合わせた形で始まる。そして同じ形のお椀をひっくり返したものが続く。これは時計の針が３を指している状態からスタートして、反時計回りにぐるっと一周した状態に対応する。針が３時の時の$y$(高さ)は$0$。12時の時は$y$（高さ）は最大の$1$（単位円として）。この要領で針が指し示す時計の円周上に赤い点があると想像し、それに対応する角度を$0$から$2\pi$まで$sin(x)$に渡した際の$x$と$y$の点をプロットして出来た曲線のイメージ          |
| $cos(x)$  | お椀を伏せて真っ二つになったような曲線が、切った中心をスタートにして始まる。そして同じ形のお椀をひっくり返したものを真っ二つに切ったものに続き、さらにその残りが続く。最後に最初のお椀の残りに続く。これは時計の針が３を指している状態からスタートして、反時計回りにぐるっと一周した状態に対応する。針が３時の時の$x$(横の幅)は最大の$1$。12時の時は$x$（横の幅）は$0$になる。この要領で針が指し示す時計の円周上に赤い点があると想像し、それに対応する角度を$0$から$2\pi$まで$sin(x)$に渡した際の$x$と$y$の点をプロットして出来た曲線のイメージ          |
| $tan(x)$  | 時計のイメージを使うと、３時のスタート地点では角度が$0$なので$Sin(0)$と$Cos(0)$はそれぞれ$0$と$1$になる(高さ$0$、横$1$)。$tan(x)$は$Sin(0)$を$Cos(0)$で割ったものなので$0$になる。途中で高さと横が同じ値になるところがある。ここで$y=1$になる。そこを過ぎて、時計の針が１２時近くまで分母はどんどん小さくなり１２時直前では分母がめちゃくちゃ小さいので高さはめちゃくちゃおおきくなる（無限大）。１２時から９時の間は逆に$y＝oo$でスタートして、分母がどんどん増えて$0$になる。ここからまたもういちど半周期がはじまる                 |



```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.animation as animation

X = np.linspace(0, 2*np.pi, 200)
Y = np.sin(X)


fig, ax = plt.subplots(1,1)
ax.set_xlim([0, 6*np.pi])
ax.set_ylim([-1.1, 1.1])

sinegraph, = ax.plot([], [])

def marker(i):
    ax.plot(X[i-1],Y[i-1],'o',color='red',markersize=5)
    ax.plot(X[i],Y[i],'o',color='black',markersize=5)


def sine(i):
    sinegraph.set_data(X[:i],Y[:i])
    ax.plot(X[i], Y[i], 'o', color='red',markersize=5)
    if i>0:
        marker(i)

anim = animation.FuncAnimation(fig, sine, frames=400, interval=50)
plt.show()
```
