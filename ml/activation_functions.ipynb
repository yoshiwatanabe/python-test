### üîΩ Á∞°Âçò„Å™„Åæ„Å®„ÇÅÔºöReLU / Sigmoid / Tanh „ÅÆÊØîËºÉ

| Èñ¢Êï∞     | Âá∫ÂäõÁØÑÂõ≤   | ‰∏≠ÂøÉ„Åå0 | ÂãæÈÖçÊ∂àÂ§±„ÅÆ„É™„Çπ„ÇØ | ‰∏ª„Å™Áî®ÈÄî                          |
|----------|------------|---------|------------------|-----------------------------------|
| ReLU     | 0 ‰ª•‰∏ä      | ‚úï       | Â∞ë„Å™„ÅÑ           | Ê∑±Â±§Â≠¶Áøí„Å´„Åä„Åë„ÇãÊ®ôÊ∫ñÁöÑ„Å™Èö†„ÇåÂ±§     |
| Sigmoid  | 0 ÔΩû 1      | ‚úï       | Ëµ∑„Åç„ÇÑ„Åô„ÅÑ       | Âá∫ÂäõÂ±§Ôºà„Éê„Ç§„Éä„É™ÂàÜÈ°û„Å™„Å©Ôºâ        |
| Tanh     | -1 ÔΩû 1     | ‚óã       | Ëµ∑„Åì„Çä„ÅÜ„Çã       | RNN„Å™„Å©„ÅÆ‰∏≠ÈñìÂ±§ÔºàÂè§ÂÖ∏ÁöÑ„Å™Áî®ÈÄîÔºâ   |


```python
import numpy as np
import matplotlib.pyplot as plt

# Define the activation functions
def relu(x):
    return np.maximum(0, x)

def sigmoid(x):
    return 1 / (1 + np.exp(-x))

def tanh(x):
    return np.tanh(x)

# Input range
x = np.linspace(-10, 10, 1000)

# Compute function values
y_relu = relu(x)
y_sigmoid = sigmoid(x)
y_tanh = tanh(x)

# Plotting
plt.figure(figsize=(12, 6))

# ReLU
plt.subplot(1, 3, 1)
plt.plot(x, y_relu, label="ReLU", color="blue")
plt.title("ReLU")
plt.grid(True)

# Sigmoid
plt.subplot(1, 3, 2)
plt.plot(x, y_sigmoid, label="Sigmoid", color="green")
plt.title("Sigmoid")
plt.grid(True)

# Tanh
plt.subplot(1, 3, 3)
plt.plot(x, y_tanh, label="Tanh", color="red")
plt.title("Tanh")
plt.grid(True)

plt.tight_layout()
plt.show()
```

### üîΩ „Åñ„Å£„Åè„ÇäÊØîËºÉË°®ÔºàReLU‰ª•Â§ñ„ÅÆ4„Å§„ÅÆÈáçË¶Å„Å™Ê¥ªÊÄßÂåñÈñ¢Êï∞Ôºâ

| Èñ¢Êï∞Âêç     | Ë≤†„ÅÆÂá∫Âäõ | ÂãæÈÖçÊ∂àÂ§±„ÅÆÂõûÈÅø | ‰∏≠ÂøÉ„Åå0 | ÁâπÂæ¥                                   |
|------------|----------|------------------|---------|----------------------------------------|
| Leaky ReLU | ‚óãÔºàÂ∞èÔºâ  | ‚óã                | ‚úï       | ReLU„ÅÆÊ≠ªÂïèÈ°å„ÇíËªΩÊ∏õ                     |
| ELU        | ‚óã        | ‚óã                | ‚óã       | Ë≤†„ÅÆÂá∫Âäõ„ÇÇËá™ÁÑ∂„ÄÅÊªë„Çâ„Åã„Å™Êõ≤Á∑ö           |
| Swish      | ‚óã        | ‚óé                | ‚óã       | GoogleÊèêÊ°à„ÄÅÂ≠¶Áøí„ÅåÂÆâÂÆö„ÄÅÈ´òÊÄßËÉΩ„Å™ÂÇæÂêë   |
| GELU       | ‚óã        | ‚óé                | ‚óã       | TransformerÁ≥ª„ÅßÊé°Áî®„ÄÅÁµ±Ë®àÁöÑ„Å´Ê¥óÁ∑¥„Åï„Çå„Å¶„ÅÑ„Çã |


```python
import numpy as np
import matplotlib.pyplot as plt

# ÂêÑÁ®ÆÊ¥ªÊÄßÂåñÈñ¢Êï∞„ÅÆÂÆöÁæ©
def leaky_relu(x, alpha=0.01):
    return np.where(x > 0, x, alpha * x)

def elu(x, alpha=1.0):
    return np.where(x >= 0, x, alpha * (np.exp(x) - 1))

def swish(x):
    return x * (1 / (1 + np.exp(-x)))  # x * sigmoid(x)

def gelu(x):
    return 0.5 * x * (1 + np.tanh(np.sqrt(2 / np.pi) * (x + 0.044715 * x**3)))

# ÂÖ•Âäõ„ÅÆÁØÑÂõ≤
x = np.linspace(-10, 10, 1000)

# Èñ¢Êï∞„Åî„Å®„ÅÆÂá∫Âäõ„ÇíË®àÁÆó
y_leaky_relu = leaky_relu(x)
y_elu = elu(x)
y_swish = swish(x)
y_gelu = gelu(x)

# „Ç∞„É©„Éï„ÅÆÊèèÁîª
fig, axes = plt.subplots(2, 2, figsize=(12, 8))

# Leaky ReLU
axes[0, 0].plot(x, y_leaky_relu, color="purple")
axes[0, 0].set_title("Leaky ReLU")
axes[0, 0].grid(True)

# ELU
axes[0, 1].plot(x, y_elu, color="orange")
axes[0, 1].set_title("ELU")
axes[0, 1].grid(True)

# Swish
axes[1, 0].plot(x, y_swish, color="green")
axes[1, 0].set_title("Swish")
axes[1, 0].grid(True)

# GELU
axes[1, 1].plot(x, y_gelu, color="blue")
axes[1, 1].set_title("GELU")
axes[1, 1].grid(True)

plt.tight_layout()
plt.show()
```
